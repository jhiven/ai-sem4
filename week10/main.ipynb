```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('milk.csv')
train_data = np.array(dataset)[:, 0:-1]
train_label = np.array(dataset)[:, -1]

train_data
```

```python
train_label
```

```python
sc = MinMaxScaler(feature_range=(0,1))
train_data = sc.fit_transform(train_data)
```

```python
train_data
```

```python
KNN = KNeighborsClassifier(n_neighbors=3, weights='distance')
KNN.fit(train_data, train_label)
```

```python
ph = float(input('Masukkan input ph:'))
temp = int(input('Masukkan input temp:'))
taste = int(input('Masukkan input taste:'))
odor = int(input('Masukkan input odor:'))
fat = int(input('Masukkan input fat:'))
turbidity = int(input('Masukkan input turbidity:'))
colour = int(input('Masukkan input colour:'))

test_data = np.array([ph, temp, taste, odor, fat, turbidity, colour])
test_data = sc.transform(np.reshape(test_data, (1, -1)))

test_data
```

```python
hasil = KNN.predict(test_data)
hasil
```

```python
dataset_training = pd.read_csv('milk_training.csv')
training_train_data = np.array(dataset_training)[:, 0:-1]
training_train_label = np.array(dataset_training)[:, -1]
```

```python
training_train_data
```

```python
training_train_label
```

```python
training_train_data = sc.fit_transform(training_train_data)
training_train_data
```

```python
dataset_testing = pd.read_csv('milk_testing.csv')
testing_test_data = np.array(dataset_testing)[:, :-1]
testing_test_label = np.array(dataset_testing)[:, -1]
```

```python
testing_test_data
```

```python
testing_test_label
```

```python
testing_test_data = sc.transform(testing_test_data)
testing_test_data
```

```python
KNN.fit(training_train_data, training_train_label)

predictions = KNN.predict(testing_test_data)
accuracy = accuracy_score(testing_test_label, predictions)
print("Accuracy:", accuracy)
```
