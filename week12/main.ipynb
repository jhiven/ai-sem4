```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
```

```python
dataset = pd.read_csv('heart.csv')
dataset
```

```python
data = dataset.loc[:,['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal']]
data
```

```python
sc = MinMaxScaler()
data_scaled = sc.fit_transform(data)
data_scaled
```

```python
kmeans = KMeans(n_clusters=2)
clusters = kmeans.fit_predict(data_scaled)
clusters
```

```python
ac = AgglomerativeClustering(n_clusters=2, linkage='single')
clusters_single = ac.fit_predict(data_scaled)

clusters_single
```

```python
ac = AgglomerativeClustering(n_clusters=2, linkage='complete')
clusters_comp = ac.fit_predict(data_scaled)

clusters_comp
```

```python
ac = AgglomerativeClustering(n_clusters=2, linkage='average')
clusters_avg = ac.fit_predict(data_scaled)

clusters_avg
```

```python
data.plot(x='age', y='sex', kind='scatter', c=clusters_avg, colormap='Paired')
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

chi2_selector = SelectKBest(chi2, k=3)
chi2_selector.fit(data, np.array(dataset)[:, -1])

cols = chi2_selector.get_support(indices=True)
df_selected_features = data.iloc[:,cols]
df_selected_features.head()
```

```python
best_attributes_col = ['thalach', 'oldpeak', 'ca']
nilai_min = None

for i in range(10):
    data_dipilih = dataset[best_attributes_col].values
    clustering = KMeans(n_clusters=3, n_init=10, random_state=i)
    cluster = clustering.fit_predict(data_dipilih)
    print(f'{i  + 1}. sse: {clustering.inertia_}')

    if nilai_min is None or nilai_min > clustering.inertia_:
        nilai_min = clustering.inertia_
```

```python
print('hasil nilai min cluster: ', nilai_min)
```
